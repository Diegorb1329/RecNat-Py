# Obtención de variable bioclimáticas (en desarrollo)

```python
import pandas as pd

# Definición de la función para extraer datos climatológicos
def extraer_datos_climatologicos(file_path):
    with open(file_path, 'r', encoding='ISO-8859-1') as file:
        # Leer las líneas del archivo
        lines = file.readlines()

    # Encontrar el inicio de los datos climatológicos
    inicio_datos = None
    for i, line in enumerate(lines):
        if line.strip().startswith('Fecha'):
            inicio_datos = i + 1
            break

    if inicio_datos is None:
        return pd.DataFrame()  # Retornar un DataFrame vacío si no se encuentra el inicio de los datos

    # Extraer y procesar los datos climatológicos
    data = []
    for line in lines[inicio_datos:]:
        parts = line.split()
        if len(parts) >= 5:  # Asegurarse de que la línea tiene suficientes partes para los datos
            fecha = parts[0]
            precip = parts[1]
            evap = parts[2] if parts[2] != 'Nulo' else None
            tmax = parts[3] if parts[3] != 'Nulo' else None
            tmin = parts[4] if parts[4] != 'Nulo' else None
            data.append([fecha, precip, evap, tmax, tmin])

    # Convertir a DataFrame
    df = pd.DataFrame(data, columns=['Fecha', 'Precip', 'Evap', 'Tmax', 'Tmin'])
    return df

# Cargar el nuevo archivo de datos climatológicos
file_path_new = '/content/RecNat-Py/Salidas de prueba/datos_crudos/dia15170.TXT'

# Utilizar la función ya definida para cargar los datos del nuevo archivo
df_new = extraer_datos_climatologicos(file_path_new)

# Convertir los tipos de datos para asegurar que las operaciones numéricas se puedan realizar
df_new['Tmax'] = pd.to_numeric(df_new['Tmax'], errors='coerce')
df_new['Tmin'] = pd.to_numeric(df_new['Tmin'], errors='coerce')
df_new['Precip'] = pd.to_numeric(df_new['Precip'], errors='coerce')

# Asegurar que las fechas estén en formato datetime correctamente
df_new['Fecha'] = pd.to_datetime(df_new['Fecha'], errors='coerce')

# Eliminar filas con valores faltantes en 'Precip', 'Tmax', o 'Tmin' después de la conversión
df_new = df_new.dropna(subset=['Precip', 'Tmax', 'Tmin'])

# Mostrar información básica del DataFrame para confirmar carga y formato correctos
df_new.info(), df_new.head()
```

```python
# Funciones auxiliares para el cálculo de variables bioclimáticas usando datos completos
# Descripciones de variables bioclimáticas
descriptions = {
    'BIO1': "Temperatura media anual.",
    'BIO2': "Rango diurno medio (diferencia entre la temperatura máxima y mínima mensual promedio).",
    'BIO3': "Isothermality (relación entre el rango diurno medio del mes y la oscilación anual de temperatura, multiplicado por 100).",
    'BIO4': "Estacionalidad de la temperatura (desviación estándar de la temperatura semanal multiplicada por 100).",
    'BIO5': "Temperatura máxima del mes más cálido.",
    'BIO6': "Temperatura mínima del mes más frío.",
    'BIO7': "Rango anual de temperatura (diferencia entre BIO5 y BIO6).",
    'BIO8': "Temperatura media del trimestre más húmedo.",
    'BIO9': "Temperatura media del trimestre más seco.",
    'BIO10': "Temperatura media del trimestre más cálido.",
    'BIO11': "Temperatura media del trimestre más frío.",
    'BIO12': "Precipitación anual total.",
    'BIO13': "Precipitación del mes más húmedo.",
    'BIO14': "Precipitación del mes más seco.",
    'BIO15': "Estacionalidad de la precipitación (coeficiente de variación).",
    'BIO16': "Precipitación del trimestre más húmedo.",
    'BIO17': "Precipitación del trimestre más seco.",
    'BIO18': "Precipitación del trimestre más cálido.",
    'BIO19': "Precipitación del trimestre más frío."
}
```

```python
def prepare_data(df, date_col):
    """Prepara los datos extrayendo mes, trimestre y año de la fecha."""
    df[date_col] = pd.to_datetime(df[date_col])
    df['Month'] = df[date_col].dt.month
    df['Quarter'] = df[date_col].dt.quarter
    df['Year'] = df[date_col].dt.year
    return df

def filter_years_with_insufficient_data(df):
    """Filtra los años con más del 10% de datos faltantes en precipitaciones."""
    # Contar total de días y días faltantes por año
    total_days_per_year = df.groupby('Year').size()
    missing_days_per_year = df[df['Precip'].isnull()].groupby('Year').size()

    # Calcular el porcentaje de días faltantes
    missing_percentage = (missing_days_per_year / total_days_per_year) * 100

    # Identificar años con menos del 10% de datos faltantes
    valid_years = missing_percentage[missing_percentage <= 10].index

    # Filtrar el DataFrame para incluir solo los años válidos
    return df[df['Year'].isin(valid_years)]

def calculate_annual_precipitation(df):
    """Calcula la precipitación total anual y luego promedia estos totales a través de los años válidos."""
    annual_precipitation = df.groupby('Year')['Precip'].sum()
    average_annual_precipitation = annual_precipitation.mean()
    return average_annual_precipitation

def calculate_bioclimatic_variables(df):
    df = filter_years_with_insufficient_data(df)  # Filtrar años con datos insuficientes
    monthly_avg_temp, quarterly_avg_temp = calculate_monthly_and_quarterly_averages(df)
    max_precip_avg, min_precip_avg, warmest_quarter_precip, coldest_quarter_precip = calculate_precipitation_extremes(df)
    average_annual_precipitation = calculate_annual_precipitation(df)  # Usar la función para calcular la precipitación anual

    results = {
        'BIO1': monthly_avg_temp[['Tmax', 'Tmin']].mean().mean(),
        'BIO2': (monthly_avg_temp['Tmax'] - monthly_avg_temp['Tmin']).mean(),
        'BIO3': 100 * ((monthly_avg_temp['Tmax'] - monthly_avg_temp['Tmin']).mean() / (monthly_avg_temp['Tmax'].max() - monthly_avg_temp['Tmin'].min())),
        'BIO4': monthly_avg_temp.std().mean() * 100,
        'BIO5': monthly_avg_temp['Tmax'].max(),
        'BIO6': monthly_avg_temp['Tmin'].min(),
        'BIO7': monthly_avg_temp['Tmax'].max() - monthly_avg_temp['Tmin'].min(),
        'BIO12': average_annual_precipitation,  # Usar el promedio calculado
        'BIO13': df.groupby(['Year', 'Month'])['Precip'].sum().groupby('Month').max().mean(),
        'BIO14': df.groupby(['Year', 'Month'])['Precip'].sum().groupby('Month').min().mean(),
        'BIO15': (df.groupby('Month')['Precip'].std() / df.groupby('Month')['Precip'].mean() * 100).mean(),
        'BIO8': quarterly_avg_temp.loc[quarterly_avg_temp['Tmin'].idxmax()].mean(),
        'BIO9': quarterly_avg_temp.loc[quarterly_avg_temp['Tmin'].idxmin()].mean(),
        'BIO10': quarterly_avg_temp['Tmax'].max(),
        'BIO11': quarterly_avg_temp['Tmin'].min(),
        'BIO16': max_precip_avg,
        'BIO17': min_precip_avg,
        'BIO18': warmest_quarter_precip,
        'BIO19': coldest_quarter_precip
    }

    bio_df = pd.DataFrame(list(results.items()), columns=['Variable', 'Valor'])
    bio_df['Definición'] = bio_df['Variable'].apply(lambda x: descriptions[x])
    return bio_df



df = prepare_data(df_new, 'Fecha')
bioclimatic_df = calculate_bioclimatic_variables(df)
bioclimatic_df
```


## 4 Operaciones con Datos 🔢

a. Variables y Tipos de Datos

Empezaremos creando variables que representen distintos tipos de datos, operadores, estructuras de control y funciones. Esto puede sonar complicado pero si se presta atención es bastante intuitivo. Puedes encontrar una extensa guía con descripciones y ejemplos [Aquí](https://www.freecodecamp.org/espanol/news/el-manual-de-python/#functionsinpython)

```python
# a. Variables y Tipos de Datos

num_arboles = 100  # int: Número de árboles en el bosque
altura_promedio = 15.0  # float: Altura promedio de los árboles en metros
tipo_bosque = "pinaceas"  # str: Tipo de bosque
alturas = []  # list: Lista para almacenar las alturas de los árboles

(num_arboles, altura_promedio, tipo_bosque,
 )
```

```python
# b. Operadores
# Ejemplo: Aumentar la altura promedio y verificar si es mayor que 15

altura_promedio += 2  # Aumentamos la altura promedio en 2 metros
altura_supera_15 = altura_promedio > 15  # Verificamos si la altura es mayor que 15 (TRUE / FALSE)

(altura_promedio, altura_supera_15) ## Verificar la nueva altura promedio
```

```python
# c. Estructuras de Control
# Ejemplo: Bucle for para generar datos aleatorios de altura de árboles

import random  # Importamos la librería random para generar números aleatorios

for _ in range(num_arboles):  # Bucle que se repite `num_arboles` veces
    altura = round(random.gauss(altura_promedio, 2), 2)  # Generamos un valor aleatorio con distribución normal
    alturas.append(altura)  # Añadimos la altura generada a la lista `alturas`

alturas[:5] ## Ver las primeras 5 alturas
```

```python
def altura_maxima(lista_alturas):
    """
    Esta función entrega la altura máxima entre los árboles.

    Parámetros:
    - lista_alturas (list): Una lista con las alturas de los árboles.

    Retorna:
    - str: Un mensaje que indica la altura máxima.
    """
    return f"La altura máxima entre los árboles es: {max(lista_alturas)} metros"
```

```python
# e. LLamado a la función

maxima_altura = altura_maxima(alturas) # Crear una variable que almacene la función con nuestro parámetro

# Resultados para verificación
(maxima_altura)  # Imprimir el resultado
```


## 4. Manejo de Datos con Python 📖
Utilizaremos [Pandas](https://aprendeconalf.es/docencia/python/manual/pandas/), [Matplotlib](https://aprendeconalf.es/docencia/python/manual/matplotlib/) y [Numpy](https://www.freecodecamp.org/espanol/news/la-guia-definitiva-del-paquete-numpy-para-computacion-cientifica-en-python/) ,  poderosas librerías de Python para la manipulación y análisis de datos. A continuación, un ejemplo básico de cómo cargar y visualizar un conjunto de datos.


```python

import pandas as pd  # Importamos la librería Pandas y podremos llamnarla si escribimos "pd" esto
                    #se puede cambiar a gusto
import numpy as np


# Generando datos aleatorios
np.random.seed(0)  # Semilla para reproducibilidad
num_muestras = 100  # Número de muestras de suelo

# Creando un DataFrame con Pandas
datos_suelo = pd.DataFrame({
    'pH': np.random.normal(6.5, 0.5, num_muestras).round(2),
    'Materia_Organica': np.random.uniform(2, 10, num_muestras).round(2),
    'Profundidad': np.random.randint(30, 100, num_muestras)
})

# Exploración de datos básicos
primeras_filas = datos_suelo.head()  # Visualizando las primeras 5 filas usando "head"
descripcion = datos_suelo.describe()  # Descripción estadística de las variables usando "describe"

(primeras_filas, descripcion)

```

La visualización de datos es muy importante. Existen diversas formas de presentarlos según la naturaleza de los datos. Puedes encontrar más imformación sobre visualización de datos [aquí](https://www.jmp.com/es_cl/statistics-knowledge-portal/exploratory-data-analysis/types-of-graphs.html#1add9be7-cde0-4d19-bad9-86f982548601)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración del estilo de los gráficos con SEABORN
sns.set_style("whitegrid")

# Histograma de pH
plt.figure(figsize=(8, 6))
sns.histplot(datos_suelo['pH'], bins=15, kde=True)
plt.title('Histograma de pH del Suelo')
plt.xlabel('pH')
plt.ylabel('Frecuencia')
plt.show()

# Diagrama de Dispersión entre pH y Materia Orgánica
plt.figure(figsize=(8, 6))
sns.scatterplot(x='pH', y='Materia_Organica', data=datos_suelo)
plt.title('Diagrama de Dispersión entre pH y Materia Orgánica')
plt.xlabel('pH')
plt.ylabel('Materia Orgánica (%)')
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
gdf_new = gpd.read_file("/content/drive/MyDrive/Libretas GC - IRNR V1/Archivos/SHP_Textura_Mezt/EDA_METZ_250000_actualizado.shp")

# Visualizar las primeras filas del nuevo GeoDataFrame
print(gdf_new.head())

```

```python
# Calcular el área de cada polígono y añadir esta información al GeoDataFrame
gdf_new['Area'] = gdf_new['geometry'].area

# Agrupar por la variable 'TEXTURA' y calcular la media del área
average_area_by_texture = gdf_new.groupby('TEXTURA')['Area'].mean()
print("\nÁrea promedio por textura del suelo:")
print(average_area_by_texture)

```

```python
# Visualización de los datos espaciales
gdf_new.plot(column='TEXTURA', cmap='viridis', legend=True, figsize=(10, 10))
plt.title('Distribución de Texturas del Suelo')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.show()

# Configurar el estilo de los gráficos
sns.set_style("whitegrid")

# Crear un boxplot para visualizar la distribución del área por textura del suelo
plt.figure(figsize=(10, 6))
sns.boxplot(x='TEXTURA', y='Area', data=gdf_new)
plt.title('Distribución del Área por Textura del Suelo')
plt.xlabel('Textura del Suelo')
plt.ylabel('Área')
plt.show()
```

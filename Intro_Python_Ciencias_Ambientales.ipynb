
## 4. Manejo de Datos con Python 游닀
Utilizaremos [Pandas](https://aprendeconalf.es/docencia/python/manual/pandas/), [Matplotlib](https://aprendeconalf.es/docencia/python/manual/matplotlib/) y [Numpy](https://www.freecodecamp.org/espanol/news/la-guia-definitiva-del-paquete-numpy-para-computacion-cientifica-en-python/) ,  poderosas librer칤as de Python para la manipulaci칩n y an치lisis de datos. A continuaci칩n, un ejemplo b치sico de c칩mo cargar y visualizar un conjunto de datos.


```python

import pandas as pd  # Importamos la librer칤a Pandas y podremos llamnarla si escribimos "pd" esto
                    #se puede cambiar a gusto
import numpy as np


# Generando datos aleatorios
np.random.seed(0)  # Semilla para reproducibilidad
num_muestras = 100  # N칰mero de muestras de suelo

# Creando un DataFrame con Pandas
datos_suelo = pd.DataFrame({
    'pH': np.random.normal(6.5, 0.5, num_muestras).round(2),
    'Materia_Organica': np.random.uniform(2, 10, num_muestras).round(2),
    'Profundidad': np.random.randint(30, 100, num_muestras)
})

# Exploraci칩n de datos b치sicos
primeras_filas = datos_suelo.head()  # Visualizando las primeras 5 filas usando "head"
descripcion = datos_suelo.describe()  # Descripci칩n estad칤stica de las variables usando "describe"

(primeras_filas, descripcion)

```

La visualizaci칩n de datos es muy importante. Existen diversas formas de presentarlos seg칰n la naturaleza de los datos. Puedes encontrar m치s imformaci칩n sobre visualizaci칩n de datos [aqu칤](https://www.jmp.com/es_cl/statistics-knowledge-portal/exploratory-data-analysis/types-of-graphs.html#1add9be7-cde0-4d19-bad9-86f982548601)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci칩n del estilo de los gr치ficos con SEABORN
sns.set_style("whitegrid")

# Histograma de pH
plt.figure(figsize=(8, 6))
sns.histplot(datos_suelo['pH'], bins=15, kde=True)
plt.title('Histograma de pH del Suelo')
plt.xlabel('pH')
plt.ylabel('Frecuencia')
plt.show()

# Diagrama de Dispersi칩n entre pH y Materia Org치nica
plt.figure(figsize=(8, 6))
sns.scatterplot(x='pH', y='Materia_Organica', data=datos_suelo)
plt.title('Diagrama de Dispersi칩n entre pH y Materia Org치nica')
plt.xlabel('pH')
plt.ylabel('Materia Org치nica (%)')
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
gdf_new = gpd.read_file("/content/drive/MyDrive/Libretas GC - IRNR V1/Archivos/SHP_Textura_Mezt/EDA_METZ_250000_actualizado.shp")

# Visualizar las primeras filas del nuevo GeoDataFrame
print(gdf_new.head())

```

```python
# Calcular el 치rea de cada pol칤gono y a침adir esta informaci칩n al GeoDataFrame
gdf_new['Area'] = gdf_new['geometry'].area

# Agrupar por la variable 'TEXTURA' y calcular la media del 치rea
average_area_by_texture = gdf_new.groupby('TEXTURA')['Area'].mean()
print("\n츼rea promedio por textura del suelo:")
print(average_area_by_texture)

```

```python
# Visualizaci칩n de los datos espaciales
gdf_new.plot(column='TEXTURA', cmap='viridis', legend=True, figsize=(10, 10))
plt.title('Distribuci칩n de Texturas del Suelo')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.show()

# Configurar el estilo de los gr치ficos
sns.set_style("whitegrid")

# Crear un boxplot para visualizar la distribuci칩n del 치rea por textura del suelo
plt.figure(figsize=(10, 6))
sns.boxplot(x='TEXTURA', y='Area', data=gdf_new)
plt.title('Distribuci칩n del 츼rea por Textura del Suelo')
plt.xlabel('Textura del Suelo')
plt.ylabel('츼rea')
plt.show()
```
